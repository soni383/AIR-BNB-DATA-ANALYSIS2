9.What is the average amount paid per customer segment?


```python
query=""" SELECT customertype,
       SUM(amountpaid) AS total_amount_paid
FROM AirbnbBookings
GROUP BY customertype
ORDER BY total_amount_paid DESC;"""
cur.execute(query)
data=cur.fetchall()
data
```

10.

Which room types generate the most revenue?


```python
query=""" select reservedroomtype,
    SUM(amountpaid) as revenue
from AirbnbBookings
group by reservedroomtype
order by revenue desc """
cur.execute(query)
data=cur.fetchall()
data

```

11.Is revenue affected by booking changes, waiting list days, or special requests?

```python
query="""SELECT bookingchanges,
       AVG(amountpaid) AS avg_revenue,
       SUM(amountpaid) AS total_revenue,
       COUNT(*) AS total_bookings
FROM AirbnbBookings
GROUP BY bookingchanges
ORDER BY bookingchanges;"""
cur.execute(query)
data=cur.fetchall()
data= pd.DataFrame(data, columns=["bookingchanges", 
                                 "avg_revenue", 
                                 "total_revenue", 
                                 "total_bookings"])

```

13.What % of guests require parking or special requests?

```python
query="""SELECT 
    (COUNT(CASE 
              WHEN requiredcarparkingspaces > 0 OR specialrequests > 0 
              THEN 1 
           END) * 100.0 / COUNT(*)) AS percentage_guests_with_parking_or_requests
FROM AirbnbBookings;"""
cur.execute(query)
data=cur.fetchall()
data
```
